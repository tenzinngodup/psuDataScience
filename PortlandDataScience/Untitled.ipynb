```python
from datetime import datetime

def formatDatetime(x): 
    return datetime.strptime(x, '%Y-%m-%d')
df['newDT'] = df['compiled from Yahoo! Finance data by Matt Borthwick'].apply(formatDatetime)
```

```python
# import matplotlib
# matplotlib.pyplot.plot_date(df['newDT'], df['ED'])

df.plot(x="newDT", y=["ED",'DD','CVX','FL','CAT'], kind="line")
```

```python
# import matplotlib
# matplotlib.pyplot.plot_date(df['newDT'], df['ED'])
import numpy as np
df['newED'] = df['ED'].apply(np.log)
df['newDD'] = df['DD'].apply(np.log)
df['newCVX'] = df['CVX'].apply(np.log)
df['newFL'] = df['FL'].apply(np.log)
df['newCAT'] = df['CAT'].apply(np.log)

df.plot(x="newDT", y=["newED",'newDD','newCVX','newFL','newCAT'], kind="line")
```
